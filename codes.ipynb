## Let's Join the dataframe to make one single dataframe

```python
respondents.merge(cities,on='City_ID',how='left')
```

```python
respondents.merge(cities,on='City_ID',how='left').merge(survey,on='Respondent_ID',how='left')
```

```python
respondents.merge(cities,on='City_ID',how='left').merge(survey,on='Respondent_ID',how='left').columns
```

## Make a new dataframe

```python
df = respondents.merge(cities,on='City_ID',how='left').merge(survey,on='Respondent_ID',how='left')
```

## Let's find some Demographic Insights like
## Who prefers energy drink more? (male/female/non-binary?)
### To find it out we're gonna focus on two colums 'Gender' and 'Consume_frequency'

```python
df['Consume_frequency'].value_counts()
```

## We can get rid of the people who drinks rarely

```python
df_drinks_more = df.query("Consume_frequency!='Rarely'").copy()
```

```python
df_drinks_more.head()
```

## Using groupby to get in general count

```python
df_drinks_more.groupby(['Gender'])['Consume_frequency'].count()
```

## According to this process males consume more energy drinks

```python
df_drinks_more.groupby(['Gender'])['Consume_frequency'].count().sort_values(ascending=False)
```

## To be more presize we can assign a weight accoring to the consumption frequency

```python
weights = {
    'Daily': 30,
    '2-3 times a week': 12,
    'Once a week': 4,
    '2-3 times a month': 3,
}
```

## Now we can create a new column and map these weights to the 'Consume_frequency' column

```python
df_drinks_more['Consume_weight'] = df['Consume_frequency'].map(weights)
```

## Let's check it out

```python
df_drinks_more.head()
```

## Now if we apply the same groupby to get the counts

```python
df_drinks_more.groupby('Gender')['Consume_weight'].sum().sort_values(ascending=False)
```

## So we'll store the first element

```python
gender_preffers_more_energy_drinks = df_drinks_more.groupby('Gender')['Consume_weight'].sum().sort_values(ascending=False).index[0]
```

```python
print("The gender that prefers energy drinks the most is:",gender_preffers_more_energy_drinks)
```

## Let's find out what age group prefers energy drinks more, we can use the same dataframe

```python
df_drinks_more.head()
```

```python
df_drinks_more.groupby(['Age'])['Consume_weight'].sum().sort_values(ascending=False)
```

## Finally to the age group that consumes more enery drinks

```python
age_group_consume_more = df_drinks_more.groupby(['Age'])['Consume_weight'].sum().sort_values(ascending=False).index[0]
```

```python
print("The age group that consume more energy drinks is:",age_group_consume_more)
```

## Let's find out what type of marketing reaches the most Youth (15-30)
