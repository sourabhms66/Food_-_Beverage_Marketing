## What are the cities the business needs to focus more on?

```python
df.head()
```

```python
df.columns
```

```python
Consumption_weights = {'Daily': 30, '2-3 times a week': 12, 'Once a week': 4, '2-3 times a month': 3,'Rarely': 1}
```

```python
df['Consume_weights'] = df['Consume_frequency'].map(Consumption_weights)
```

```python
df.head()
```

```python
df.groupby('City')['Consume_weights'].mean().sort_values()
```

```python
average_consumption = df['Consume_weights'].mean()
```

```python
brand_perception_weights = {'Positive': 1, 'Neutral': 0, 'Negative': -1}
```

```python
df['Brand_perception_weights'] = df['Brand_perception'].map(brand_perception_weights)
```

```python
df.head()
```

```python
df.groupby('City')['Brand_perception_weights'].mean().sort_values()
```

```python
df.groupby('City').agg({
    'Brand_perception_weights': 'mean',
    'Consume_frequency': 'value_counts'  # This gives you frequency distribution by city
})
```

```python
target_cities = df.groupby('City').agg({
    'Brand_perception_weights': 'mean',
    'Consume_frequency': 'value_counts'  # This gives you frequency distribution by city
})
```

```python
# Sort by brand perception to identify cities with lower scores
city_focus_sorted = target_cities.sort_values('Brand_perception_weights')

# Output the cities that need focus
print("Cities that need more focus based on brand perception:")
print(city_focus_sorted)
```

```python
# Adjust the threshold as per your requirement
threshold_perception = 0  # Neutral
threshold_consumption = df['Consume_frequency'].map(Consumption_weights).mean()  # Average consumption score

low_perception_cities = df[df['Brand_perception_weights'] <= threshold_perception]
low_consumption_cities = df[df['Consume_frequency'].map(Consumption_weights) <= threshold_consumption]

# Combining the filters to find cities needing more focus
cities_needing_focus = low_perception_cities.merge(low_consumption_cities, on='City', how='inner')['City'].unique()

print("Cities that need more focus based on both low brand perception and low consumption frequency:")
print(cities_needing_focus)
```
