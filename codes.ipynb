## Make a new dataframe

```python
df = respondents.merge(cities,on='City_ID',how='left').merge(survey,on='Respondent_ID',how='left')
```

## Let's find some Demographic Insights
## Who prefers energy drink more? (male/female/non-binary?)
### To find it out we're gonna focus on two colums 'Gender' and 'Consume_frequency'

```python
df['Consume_frequency'].value_counts()
```

## We can get rid of the people who drinks rarely

```python
df_drinks_more = df.query("Consume_frequency!='Rarely'").copy()
```

```python
df_drinks_more.head()
```

## Using groupby to get in general count

```python
df_drinks_more.groupby(['Gender'])['Consume_frequency'].count()
```

## According to this process males consume more energy drinks

```python
df_drinks_more.groupby(['Gender'])['Consume_frequency'].count().sort_values(ascending=False)
```

## To be more presize we can assign a weight accoring to the consumption frequency

```python
weights = {
    'Daily': 30,
    '2-3 times a week': 12,
    'Once a week': 4,
    '2-3 times a month': 3,
}
```

## Now we can create a new column and map these weights to the 'Consume_frequency' column

```python
df_drinks_more['Consume_weight'] = df['Consume_frequency'].map(weights)
```
